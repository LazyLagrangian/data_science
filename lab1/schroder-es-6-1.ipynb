```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def factorial(n):
    if n != 0:
        n *= factorial(n-1)
        return n
    else:
        return 1
```

```python
fig , ax = plt.subplots(2)
N = 100
q = 500
x = np.linspace(1, q, q)
y = np.zeros(q)

#for i in range(500):
for i in range(20):
    microstates = factorial((q-i-1)+N-1)/factorial((q-i-1))/factorial(N-1)
    y[i] = microstates


#ax[0].plot(x,y)
ax[0].plot(x[:20],y[:20])
ax[1].plot(x, np.log(y))
```

interesting that when plotting till the last quanta of energy (500 instead of 20) the log of y is no longer linear, because the reservoir no longer has >> energy than the system
